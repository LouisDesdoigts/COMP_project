```python
import os
import sys
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import glob
from astropy.table import Table
from copy import deepcopy
```

```python
cat = Table.read('jvds_stelkin_cat_v012_mge_v310820_private.fits', hdu=1)
```

RE_CIRC_KPC:      FLOAT - Effective Circularised radius in KiloParsec from MGE (F. D'Eugenio in Prep.). \
MU_AT_RE:         FLOAT - Mean surface density at one Re from Bryant et al. 2015 \
LMSTAR:           FLOAT - Log_10 Stellar mass derived from g-i colour from Bryant et al. 2015 and Owers et al. 2017. \ \

GMINUSI:          FLOAT - g-i colour from from Bryant et al. 2015 and Owers et al. 2017. \
MTYPE:            FLOAT - SAMI visual morphological type from Cortese et al. 2016 \
N_SERSIC:         FLOAT - Sersic index from the GAMA best-fit 2D Sersic profile in the r-band Lee et al. 2012 (Concentration)

```python
# Load in our mass radius, mtype and ID data
radius = cat.field('RE_CIRC_KPC')
density = cat.field("MU_AT_RE")
log_mass = cat.field('LMSTAR')

colour = cat.field("GMINUSI")
mtype = cat.field("MTYPE")
sindex = cat.field("N_SERSIC")
```

```python
# This is done in order to make the visualisation of the morphological types easier through a colorbar function 

mtype_old = deepcopy(mtype)
for i in range(len(mtype_old)):
    if mtype_old[i] == -9:
        mtype[i] = 6
```

Initial exploration of the distribution of each data set to look for outliers

```python
data_vectors = [radius, density, log_mass, colour, mtype, sindex]
data_names = ["radius", "density", "log_mass", "colour", "mtype", "sindex"]
```

```python
n = len(data_vectors)

plt.figure(figsize=(15, 10))

for i in range(n):
    plt.subplot(2, n//2, i+1)
    plt.hist(data_vectors[i])
    plt.title("{}".format(data_names[i]))
    plt.xlabel(data_names[i])
        
plt.tight_layout()
```

clearly we need to clean some outliers

```python
# Get rid of the crazy outliers, (some crazy sindex one, and the -9 and 5 entires of morphological type)
m1 = sindex > 0
m2 = mtype < 5
m3 = radius <= 10
m4 = (density <= 27.5)*(density >= 18)
m5 = (colour >= 0)*(colour <= 1.5)
m6 = sindex <= 8
```

```python
mask = m1*m2*m3*m4*m5*m6
mask = m1*m3*m4*m5*m6
mask = m1

n = len(data_vectors)
plt.figure(figsize=(15, 10))

for i in range(n):
    plt.subplot(2, n//2, i+1)
    plt.hist(data_vectors[i][mask], bins=30)
    plt.title("{}".format(data_names[i]))
    plt.xlabel(data_names[i])
        
plt.tight_layout()
plt.savefig("parameter_distribution")
```

## This explores the basic relationships between each of the main continuous parameters

```python
data_small = [radius, density, log_mass, colour, sindex]
data_names_small = ["radius", "density", "log_mass", "colour", "sindex"]

plt.figure(figsize=(50, 35))
idx = 1
for i in range(n-1):
    for j in range(n-1):
        if i > j:
            plt.subplot(n-1, n-1, idx)
            plt.scatter(data_small[i][mask], data_small[j][mask], alpha=0.5)
            plt.title("{} vs {}".format(data_names_small[i], data_names_small[j]))
            plt.xlabel(data_names_small[i])
            plt.ylabel(data_names_small[j])

        idx += 1
plt.tight_layout()
# plt.savefig("parameter_relationships")
```

```python
data_small = [radius, density, log_mass, colour, sindex]
data_names_small = ["radius", "density", "log_mass", "colour", "sindex"]

plt.figure(figsize=(50, 35))
idx = 1
for i in range(n-1):
    for j in range(n-1):
        if i > j:
            plt.subplot(n-1, n-1, idx)
            c = plt.scatter(data_small[i][mask], data_small[j][mask], c=mtype[mask], alpha=0.5)
            plt.title("{} vs {}".format(data_names_small[i], data_names_small[j]))
            plt.xlabel(data_names_small[i])
            plt.ylabel(data_names_small[j])
            plt.colorbar(c)
        
        idx += 1
plt.tight_layout()
plt.savefig("parameter_relationships")
```
